# Loan Default EDA
**Dataset:** Give Me Some Credit


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed/clean_loans.csv')
print(df.shape)
df.head()
```

```python
# Target distribution
print('Default Rate:', df['SeriousDlqin2yrs'].mean())
sns.countplot(x='SeriousDlqin2yrs', data=df)
plt.title('Default Distribution')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```

```python
# Default rate by age band
df['age_band'] = pd.cut(df['age'], bins=[0,25,35,50,120], labels=['<25','25-35','35-50','50+'])
df.groupby('age_band', observed=True)['SeriousDlqin2yrs'].mean().plot(kind='bar', color='salmon')
plt.title('Default Rate by Age Group')
plt.ylabel('Default Rate')
plt.xticks(rotation=0)
plt.show()
```

```python
# Revolving utilization vs default
fig, axes = plt.subplots(1,2,figsize=(14,5))
df[df['SeriousDlqin2yrs']==0]['RevolvingUtilizationOfUnsecuredLines'].hist(ax=axes[0], bins=50, alpha=0.7, label='No Default', color='steelblue')
df[df['SeriousDlqin2yrs']==1]['RevolvingUtilizationOfUnsecuredLines'].hist(ax=axes[1], bins=50, alpha=0.7, label='Default', color='red')
axes[0].set_title('No Default - Utilization')
axes[1].set_title('Default - Utilization')
plt.tight_layout()
plt.show()
```
